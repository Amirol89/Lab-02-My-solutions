```python

k = 3.0   # replace with value shown in your instructions
c = 4.50  # replace with value shown in your instructions

f = np.sqrt(k * x)
g = (x - c)**2

plt.figure()
plt.plot(x, f, label='f(x)')
plt.plot(x, g, label='g(x)')
plt.title("Overview of both functions")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```
