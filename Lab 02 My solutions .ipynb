### My solution

```python
x = np.linspace(0, 10, 500)

k = 4.2
c = 4.81

f = np.sqrt(k * x)
g = (x - c)**2
```

```python
plt.figure()
plt.plot(x, f, label="f(x) = sqrt(4.2x)")
plt.plot(x, g, label="g(x) = (x - 4.81)^2")

plt.title("Overview: Both functions")
plt.xlabel("x")
plt.ylabel("y")

plt.legend()
plt.grid(True)

plt.show()
```

```python
plt.figure()
plt.plot(x, f, label="f(x)")
plt.plot(x, g, label="g(x)")

plt.xlim(0, 3)
plt.ylim(0, 4)

plt.title("Intersection 1 (zoomed)")
plt.xlabel("x")
plt.ylabel("y")

plt.legend()
plt.grid(True)

plt.show()
```

```python
plt.figure()
plt.plot(x, f, label="f(x)")
plt.plot(x, g, label="g(x)")

plt.xlim(4, 7)
plt.ylim(0, 8)

plt.title("Intersection 2 (zoomed)")
plt.xlabel("x")
plt.ylabel("y")

plt.legend()
plt.grid(True)

plt.show()
```
